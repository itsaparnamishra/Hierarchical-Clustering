```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
#read the file 
uni = pd.read_csv("Universities_Clustering.csv")
uni.head(10)
```

```python
#normalise or standardise the data 
#defining function 
def std_func(i):
    z = (i - np.mean(i)) / np.std(i)
    return z 
```

```python
df_standard = std_func(uni.iloc[:,1:])
df_standard
```

```python
# standardise the data using libraries 


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
uni_data = pd.DataFrame(scaler.fit_transform(uni.iloc[:,1:]))
uni_data
```

## we can also normalise our data

```python
#defining function 

def norm_func(i):
    n = (i - np.min(i))/(np.max(i) - np.min(i))
    return n
```

```python
df_norm = norm_func(uni.iloc[:,1:])
df_norm
```

```python
# we can also normalise using library

from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(uni.iloc[:,1:]))
data
```

```python
# find linkages 

from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch

z = linkage(df_standard, method = "complete",metric = "euclidean")
plt.figure(figsize = (15,5))
plt.title("Hierarchical clustering Dendogram")
plt.xlabel("Observations")
plt.ylabel("Distance")
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(uni_data) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
```

```python
uni['clust']=cluster_labels # creating a  new column and assigning it to new column 
uni
```
